```python
import os, random, shutil

image_dir = "dataset/images"
label_dir = "dataset/labels"

train_img = "dataset/images/train"
val_img = "dataset/images/val"
train_lbl = "dataset/labels/train"
val_lbl = "dataset/labels/val"

os.makedirs(train_img, exist_ok=True)
os.makedirs(val_img, exist_ok=True)
os.makedirs(train_lbl, exist_ok=True)
os.makedirs(val_lbl, exist_ok=True)

images = [img for img in os.listdir(image_dir) if img.endswith((".jpg", ".png"))]
random.shuffle(images)

split = int(0.8 * len(images))

for i, img in enumerate(images):
    label = os.path.splitext(img)[0] + ".txt"

    if not os.path.exists(f"{label_dir}/{label}"):
        continue

    if i < split:
        shutil.copy(f"{image_dir}/{img}", f"{train_img}/{img}")
        shutil.copy(f"{label_dir}/{label}", f"{train_lbl}/{label}")
    else:
        shutil.copy(f"{image_dir}/{img}", f"{val_img}/{img}")
        shutil.copy(f"{label_dir}/{label}", f"{val_lbl}/{label}")

print("✅ Split Done in Colab")
```

```python
import os
from collections import Counter

image_dir = "/content/dataset/images"
label_dir = "/content/dataset/labels"

image_exts = (".jpg", ".png", ".jpeg")

images = [f for f in os.listdir(image_dir) if f.endswith(image_exts)]
labels = [f for f in os.listdir(label_dir) if f.endswith(".txt")]

print(f"📸 Total Images: {len(images)}")
print(f"🏷 Total Labels: {len(labels)}")

# 🔍 Missing labels
missing_labels = []
for img in images:
    label_name = os.path.splitext(img)[0] + ".txt"
    if label_name not in labels:
        missing_labels.append(img)

print(f"❌ Images without labels: {len(missing_labels)}")

# 🔍 Missing images
missing_images = []
for lbl in labels:
    img_name = os.path.splitext(lbl)[0] + ".jpg"
    if img_name not in images:
        missing_images.append(lbl)

print(f"❌ Labels without images: {len(missing_images)}")

# 🔍 Empty labels
empty_labels = []
for lbl in labels:
    path = os.path.join(label_dir, lbl)
    if os.path.getsize(path) == 0:
        empty_labels.append(lbl)

print(f"⚠ Empty label files: {len(empty_labels)}")

# 🔍 Class distribution
class_counter = Counter()

for lbl in labels:
    path = os.path.join(label_dir, lbl)
    with open(path, "r") as f:
        lines = f.readlines()
        for line in lines:
            parts = line.strip().split()
            if len(parts) == 5:
                class_id = int(parts[0])
                class_counter[class_id] += 1

print("\n📊 Class Distribution:")
for cls, count in sorted(class_counter.items()):
    print(f"Class {cls}: {count}")

# 🔍 Invalid annotations
invalid_labels = []

for lbl in labels:
    path = os.path.join(label_dir, lbl)
    with open(path, "r") as f:
        lines = f.readlines()
        for line in lines:
            parts = line.strip().split()
            if len(parts) != 5:
                invalid_labels.append(lbl)

print(f"\n❌ Invalid label format files: {len(invalid_labels)}")

print("\n✅ Dataset Check Complete!")
```

```python
import os
from collections import Counter

base_path = "dataset"

folders = ["train", "val"]
image_exts = (".jpg", ".png", ".jpeg")

for folder in folders:
    print(f"\n🔍 Checking {folder.upper()} dataset...\n")

    image_dir = os.path.join(base_path, "images", folder)
    label_dir = os.path.join(base_path, "labels", folder)

    images = [f for f in os.listdir(image_dir) if f.endswith(image_exts)]
    labels = [f for f in os.listdir(label_dir) if f.endswith(".txt")]

    print(f"📸 Total Images: {len(images)}")
    print(f"🏷 Total Labels: {len(labels)}")

    # ❌ Missing labels
    missing_labels = []
    for img in images:
        lbl = os.path.splitext(img)[0] + ".txt"
        if lbl not in labels:
            missing_labels.append(img)

    print(f"❌ Images without labels: {len(missing_labels)}")

    # ❌ Missing images
    missing_images = []
    for lbl in labels:
        img = os.path.splitext(lbl)[0] + ".jpg"
        if img not in images:
            missing_images.append(lbl)

    print(f"❌ Labels without images: {len(missing_images)}")

    # ⚠ Empty labels
    empty_labels = []
    for lbl in labels:
        path = os.path.join(label_dir, lbl)
        if os.path.getsize(path) == 0:
            empty_labels.append(lbl)

    print(f"⚠ Empty label files: {len(empty_labels)}")

    # 📊 Class distribution
    class_counter = Counter()

    for lbl in labels:
        path = os.path.join(label_dir, lbl)
        with open(path, "r") as f:
            lines = f.readlines()
            for line in lines:
                parts = line.strip().split()
                if len(parts) == 5:
                    class_id = int(parts[0])
                    class_counter[class_id] += 1

    print("\n📊 Class Distribution:")
    for cls, count in sorted(class_counter.items()):
        print(f"Class {cls}: {count}")

    # ❌ Invalid format
    invalid_labels = []

    for lbl in labels:
        path = os.path.join(label_dir, lbl)
        with open(path, "r") as f:
            lines = f.readlines()
            for line in lines:
                parts = line.strip().split()
                if len(parts) != 5:
                    invalid_labels.append(lbl)

    print(f"\n❌ Invalid label format files: {len(invalid_labels)}")

print("\n✅ FULL DATASET CHECK COMPLETE!")
```
